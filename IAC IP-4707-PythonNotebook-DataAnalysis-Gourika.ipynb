```python
#Change the datatype and numerical value of Family Income
import re

def convert_lakh_to_numeric(lakh_str):
    """
    Convert a 'Lakh' value string to its numeric equivalent.

    Parameters:
    lakh_str (str): The string containing the Lakh value.

    Returns:
    float: The numeric equivalent of the Lakh value.
    """
    # Check if the value is a range (e.g., "5-7 Lakh")
    if '-' in lakh_str:
        # Extract the two numbers
        start, end = map(int, re.findall(r'\d+', lakh_str))
        # Convert to numeric and take the average
        return (start * 100000 + end * 100000) / 2
    elif 'Lakh+' in lakh_str:
        # Handle case like "7 Lakh+"
        number = int(re.search(r'\d+', lakh_str).group())
        return number * 100000
    elif 'Lakh' in lakh_str:
        # Handle cases like "5 Lakh"
        number = int(re.search(r'\d+', lakh_str).group())
        return number * 100000
    else:
        # If the format is unexpected, return NaN or handle as needed
        return None

# Apply the function to the 'Family Income' column and update it in place
df['Family Income'] = df['Family Income'].apply(convert_lakh_to_numeric)

# Display the updated DataFrame
print(df[['Family Income']])
```

```python
# Change the numerical value of expected salary
def convert_salary_to_numeric(salary_str):
    """
    Convert an 'Expected salary (Lac)' value string to its numeric equivalent in full value.

    Parameters:
    salary_str (str): The string containing the salary value in Lac.

    Returns:
    float: The numeric equivalent of the salary value in full value.
    """
    try:
        # Check if the salary string is already numeric
        if isinstance(salary_str, (int, float)):
            return float(salary_str) * 100000

        # Extract the numeric part from the string and convert to full numeric value
        number = float(salary_str)
        return number * 100000
    except ValueError:
        # Handle cases where conversion fails
        return None

# Apply the function to the 'Expected salary (Lac)' column and update it in place
df['Expected salary (Lac)'] = df['Expected salary (Lac)'].apply(convert_salary_to_numeric)

# Display the updated DataFrame to confirm the conversion
print(df[['Expected salary (Lac)']])
```

```python
def plot_data_distribution(dataframe, bins=20):
    """
    Plot histograms for the numeric columns in the provided DataFrame to check data distribution.

    Parameters:
    dataframe (pd.DataFrame): The DataFrame containing the data to analyze.
    bins (int): The number of bins for the histograms (default is 20).

    Returns:
    None
    """
    # Select numeric columns
    numeric_columns = dataframe.select_dtypes(include=['int64', 'float64']).columns

    if not numeric_columns.any():
        print("No numeric columns found in the DataFrame.")
        return

    # Create histograms for each numeric column
    for column in numeric_columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(dataframe[column].dropna(), bins=bins, kde=False)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()

# Example usage:
plot_data_distribution(df)

```

**QUESTION 7:** Are there any outliers in the quantity (number of courses completed) attribute?

**Answer:** There are no outliers in the quantity attribute.

```python
import numpy as np

# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers in the 'Quantity' column
outliers = df[(df['Quantity'] < lower_bound) | (df['Quantity'] > upper_bound)]

print("Number of outliers in the 'Quantity' column:", outliers.shape[0])
print("Outliers in the 'Quantity' column:")
print(outliers)
```
